# Data science capstone project

### **Introduction**

This report analyses to find the best neighborhood location to a new aspiring chef to open a Chinese Restaurant in  Chennai,a city in India's Southern State.The location should have less clustering of Chinese Restaurants around it.

### **Data Section**

This report uses the Foursquare loaction data and also the kaggle dataset https://www.kaggle.com/phiitm/chennai-zomato-restaurants-data for analysis.
The kaggle dataset consists of name of Neighborhood,rating and type of restaurant.
